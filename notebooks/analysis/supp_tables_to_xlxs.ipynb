```python
import sys
import pandas as pd

sys.path.insert(0, "../..")  # add project_config to path
import project_config

supp_table_dir = project_config.SUPPLEMENTARY_TABLES_DIR

supp_table_dict = {
    "Table_S1": "joint_simulation_predictionMetrics_per_seed_grouped.csv",
    "Table_S2": "joint_simulation_predictionMetrics_per_seed.csv",
    "Table_S3": "spike_in_simulation_predictionMetrics_per_seed_grouped.csv",
    "Table_S4": "spike_in_simulation_predictionMetrics_per_seed.csv",
    "Table_S5": "spike_in_simulation_modelComparison_testSet.csv",
    "Table_S6": "p1000_predictionMetrics_per_seed_grouped.csv",
    "Table_S7": "p1000_predictionMetrics_per_seed.csv",
    "Table_S8": "p1000_predictionMetrics_per_datasplit_grouped.csv",
    "Table_S9": "p1000_predictionMetrics_per_datasplit.csv",
    "Table_S10": "p1000_datasplit_modelComparison_testSet.csv",
    "Table_S11": "p1000_datasplit_datasetComparison_testSet.csv",
    "Table_S12": "p1000_datasplit_brca2_rankComparison_testSet.csv",
    "Table_S13": "p1000_datasplit_patientBootstrap_AUPRC.csv",
    "Table_S14": "p1000_datasplit_patientBootstrap_AUPRC_model_comparison.csv",
    "Table_S15": "p1000_datasplit_patientBootstrap_AUPRC_dataset_comparison.csv",
}

output_path = supp_table_dir / "Supplementary_Tables_S1-S15.xlsx"

with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    
    # --- README sheet first ---
    readme_df = pd.DataFrame({
        "Supplementary Tables S1-S15": [
            "Each sheet corresponds to a numbered supplementary table.",
            "See supplementary material in manuscript for full captions."
        ]
    })
    readme_df.to_excel(writer, sheet_name="README", index=False)
    writer.sheets["README"].freeze_panes = "A2"

    # --- Write tables in dict order ---
    for sheet_name, csv_name in supp_table_dict.items():
        csv_path = supp_table_dir / csv_name
        
        df = pd.read_csv(csv_path)
        df.to_excel(writer, sheet_name=sheet_name, index=False)
        
        # Freeze first row (header)
        writer.sheets[sheet_name].freeze_panes = "A2"

print(f"Written: {output_path.resolve()}")
```

```python
from pathlib import Path

# All files in directory
all_files = {p.name for p in Path(supp_table_dir).glob("*.csv")}

# Files included in your dict
included_files = set(supp_table_dict.values())

# Files not included
not_included = sorted(all_files - included_files)



print("Files in directory NOT included in supp_table_dict:\n")
for f in not_included:
    print(f)
```
